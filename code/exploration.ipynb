```python
import spacy
import en_core_web_sm

nlp = spacy.load("en_core_web_sm")

def is_english(text):
    
    from langdetect import detect

    try:
        return detect(text) == 'en'
    except LangDetectException:
        return False
```

```python
def clean_text(text):    
     
    import emoji, regex as re

    text.lower()
    text = re.sub(r'\bhttp\S*\b', '', text) #clean links
    text = emoji.replace_emoji(text, replace='') #clean emojis
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text) #clean special n punctuation

    return text
```

```python
pip install tqdm
```

```python
from swifter import swifter
from pandarallel import pandarallel
from tqdm import tqdm
pandarallel.initialize(progress_bar=True)
tdqm.pandas(desc='preprocessing')
docs = yoda_csv['content'][:5].to_frame()

docs['content'] = docs['content'].parallel_apply(clean_text)
#print(docs)
temp = docs['content'].parallel_apply(is_english) #create temp df to hold 
docs = docs[temp].reset_index(drop = True)
#print('1', docs)
#cleaned_text = (clean_text(t) for t in docs['content']) 

content, tokens, strings = [], [], [] #create lists for df columns

for doc in tqdm(nlp.pipe(docs['content'],batch_size=500, n_process=4), desc="spaCy processing"): #[LOOK HERE] modify batch size and num processes for your needs
    lemma_list = [token.lemma_.lower() for token in doc if not token.is_stop]
    lemma_string = ' '.join(lemma_list).strip()

    content.append(doc.text.strip())
    tokens.append(lemma_list)
    strings.append(lemma_string)

docs = pd.DataFrame({"content": content, "tokens": tokens, "strings": strings}) #push processed text to df 
```

```python
import pyLDAvis.gensim_models
import gensim
from gensim.corpora.dictionary import Dictionary
from gensim.models import LdaMulticore
from gensim.models import CoherenceModel
import gensim.corpora as corpora
from pprint import pprint
pyLDAvis.enable_notebook()
```

```python
text = docs['tokens']
id2word = corpora.Dictionary(text) #needs list of tokens as input
corpus = [id2word.doc2bow(word) for word in text]
```

```python
num_topics = 20

#build model
lda = gensim.models.LdaMulticore(corpus = corpus, id2word = id2word, num_topics = num_topics)
pprint(lda.print_topics())
```

```python
lda_display = pyLDAvis.gensim_models.prepare(lda, corpus, id2word)
pyLDAvis.display(lda_display)
```
